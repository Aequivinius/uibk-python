## Shakespeare's favorite words

Find the 100 most common words in the corpus of Shakepeare's work.

### Get the files of the corpus

```python
import os

def path_corpus(path):
    return [os.path.join(path, f) for f in os.listdir(path) if os.path.isfile(os.path.join(path, f))]

files = path_corpus(os.path.join("", "..", "exercise-5", "corpus"))
print((files[0]))
```

### Tokenize the files and calculate the counts

```python
def tokenize_file(path):
    tokens = ""
    with open(path, 'r') as f:
        tokens = f.read()
        
    tokens_as_list = tokens.lower().split()
    normalized_tokens = []
    
    for token in tokens_as_list:
        normalized_token = ''.join(char for char in token if char.isalnum())
        if normalized_token !="":
            normalized_tokens.append(normalized_token)
    
    return normalized_tokens
```

```python
def calculate_counts(pathlist):
    counts = {}
    for path in pathlist:
        tokens = tokenize_file(path)
        
        for token in tokens:
            if token in counts:
                counts[token] = counts[token] + 1
            else:
                counts[token] = 1
    return counts

counts = calculate_counts(files)
```

### Sort the tokens

```python
def sort_counts(counts):
    sorted_collection = sorted(counts.items(), key=lambda item: item[1], reverse=True)
    return sorted_collection

sorted_counts = sort_counts(counts)
```

### Create CSV File

```python
def write_frequencies(frequencies, path):
    rank = 1
    sum_all_counts = 0
    for entry in frequencies:
        sum_all_counts = sum_all_counts + entry[1]
    
    with open(path, 'w') as f:
        for entry in frequencies:
            frequency = entry[1]/sum_all_counts
            f.write(str(rank) + "," + entry[0] + "," + str(entry[1]) + "," + str(frequency) + "\n")
            
            rank = rank + 1
    
    return
    
write_frequencies(sorted_counts, 'frequencies.csv')
```

### Print 100 most common words

```python
def print_100mostcommonwords(frequencies):
    rank = 1
    sum_all_counts = 0
    for entry in frequencies:
        sum_all_counts = sum_all_counts + entry[1]

    print("Rank\t" + "Word\t" + "Count\t")
    for entry in frequencies:
        frequency = entry[1]/sum_all_counts
        print(str(rank) + "\t" + entry[0] + "\t" + str(entry[1]) + "\t")
        rank = rank + 1
        if rank ==  101:
            break

    return

print_100mostcommonwords(sorted_counts)
```
