```python
import os
from tqdm.auto import tqdm

corpus_path = "maildir/"
file_paths = []
for (dirpath, dirnames, filenames) in os.walk(corpus_path):
    file_paths += [ os.path.join(dirpath, filename) for filename in filenames ]
```

```python
documents = []
for file_path in tqdm(file_paths[:100]):
    with open(file_path) as f:
        documents.append(nlp(f.read()))
```

```python
document = nlp("I think you are stupid")
displacy.render(document,style='dep',jupyter=True)
```

```python
for document in documents:
    for sentence in document.sents:
        for token in sentence:
            if token.head.lemma_ == "think" and token.dep_ == "ccomp":
                print(sentence.text.replace("\n",""))
                print("‚û°Ô∏è" + " ".join([t.text for t in token.subtree]).replace("\n",""))
                print()
```
