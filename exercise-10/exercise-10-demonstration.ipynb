```python
import matplotlib.pyplot as plt
```

```python
figure, ax = plt.subplots(1, 2, figsize=(20,6))
figure.suptitle("Title")
plt.show()
```

```python
figure, ax = plt.subplots(1, 4, figsize=(20,6))
figure.suptitle("4 types of plots")

# x = [0, 2]
x = ["ente", "aubergine"]
y = [1, 3]

ax[0].bar(x, y, color="tab:green")
ax[1].scatter(x, y, s=(10, 100), alpha=0.5)
ax[2].pie(y, labels=x)
ax[3].plot(x, y, linestyle='dashed', linewidth=2)
plt.show()
```

```python
import math

figure, ax = plt.subplots(1, 2, figsize=(20,6))
figure.suptitle("dots and functions")

x1 = range(0, 100)
# you would use numpy for this usually
x2 = [ x / 100 for x in range(0, 10000) ]
f1 = [ x * 2 for x in x1 ]
f2 = [ math.sin(x) for x in x2 ]
ax[0].plot(x1, f1, c="tab:blue")
ax[1].plot(x2, f2, c="tab:red")

plt.show()
```

```python
figure, ax = plt.subplots(1, 1, figsize=(20,6))
figure.suptitle("making data look nice")

x1 = range(0, 100)
# you would use numpy for this usually
x2 = [ x / 100 for x in range(0, 10000) ]
f1 = [ x * 2 for x in x1 ]
f2 = [ 100 * math.sin(x) for x in x2 ]
ax.plot(x1, f1, c="tab:blue", label="Tontaruru", linewidth=5)
ax.plot(x2, f2, c="tab:red", label="Miromente", linestyle="--")
ax.legend(loc='best')
ax.tick_params(labelcolor="tab:green", labelsize=25)
ax.grid()

plt.show()
```
